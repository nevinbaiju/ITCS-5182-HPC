```python
import math
import matplotlib.pyplot as plt
import time
```

```python

def advection(n, dt, plot=False):
    t_max = 2
    x_min = 0
    x_max = 1
    v = 1
    x_c = 0.25
    dx = (x_max-x_min)/n
    nb_steps = int(t_max/dt)
    alpha = (v*dt)/(2*dx);

    x = []
    for i in range(n+3):
        x.append(x_min + (i-1) * dx)

    u_0 = []
    for i in range(n+3):
        u_0.append(math.exp(-200*(x[i] - x_c)*x[i]-x_c))

    u = list(u_0)
    u_new = list(u_0)

    res = []
    for timestamp in range(nb_steps):
        current_time = timestamp*dt
        for j in range(n+1):
            u_new[j] = u[j] - alpha*(u[j+1] - u[j-1]) + 0.5*(u[j+1] - 2*u[j] + u[j-1])

        u = list(u_new)
        u[0] = u[n+1]
        u[n+2] = u[1]
        res.append(u[n+2])
        if timestamp%500 == 0 and plot:
            plt.figure(figsize=(20, 5))
            plt.plot(u)
            plt.title(f'timestamp={timestamp}')
            plt.show()
        elif timestamp%500 == 0:
            print(f'{timestamp}/{nb_steps}', sep='\r', end='\r')
```

```python
start_time = time.time()
advection(n=1003, dt=0.0009, plot=True)
print(time.time()-start_time)
```
