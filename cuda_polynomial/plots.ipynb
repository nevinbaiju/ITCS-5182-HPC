```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
headers = ['n', 'degree', 'flops', 'memory_bandwidth', 'pci_latency', 'pci_bandwidth']
bw_1 = pd.read_csv('results/bw_1.txt', header=None)
bw_2 = pd.read_csv('results/bw_2.txt', header=None)
bw = pd.concat([bw_1, bw_2])

flops = pd.read_csv('results/flops.txt', header=None)
pci_l = pd.read_csv('results/pci_l.txt', header=None)
bw.columns = flops.columns = pci_l.columns = headers
```

### Latency

```python
plt.hist(pci_l['pci_latency'], bins=30)
```

### flops

```python
plt.hist(flops['flops'])
```

### Bandwidth

#### PCI Bandwidth

```python
plt.figure(figsize=(12, 6))
plt.title('PCI Bandwidth for different sizes of the array')
plt.xlabel('Size of array')
plt.ylabel('PCI Bandwidth (GB/s)')
plt.plot(bw['n'], bw['pci_bandwidth'])
```

#### GPU Memory bandwidth

```python
plt.figure(figsize=(12, 6))
plt.title('GPU Memory Bandwidth for different sizes of the array')
plt.xlabel('Size of array')
plt.ylabel('GPU Memory Bandwidth (GB/s)')
plt.plot(bw['n'], bw['memory_bandwidth'])
```
